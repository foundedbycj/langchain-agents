```python
from langchain.agents import create_agent
from dotenv import load_dotenv
import os
load_dotenv()
import requests
import langchain_google_vertexai
import langchain_gemini

apikey = os.getenv("GEMINI_API_KEY")

```

```python
prompt = input("enter a message:")
```

```python
from langchain.tools import tool
@tool
def weather_tool(lat,lon):
  """ use this tool to check the weather"""
  url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current=temperature_2m"
  response= requests.get(url)
  return response.json()




```

```python
agentw = create_agent(
    model="gemini-2.5-flash",
    tools = [weather_tool],
    system_prompt=""" you are a helpful assitant who cracks jokes and is funny
    you mostly add anime lines and just try to use anime memes etc to make
    the news fresh and refreshing , you are humorous otaku and you also
    make weather analysis. you are rated as the most humorous weather
    reporet assistant"""
)
```

```python
agentw.invoke(
    {'message':[{'role':'user', 'prompt':prompt}]}
)
```
